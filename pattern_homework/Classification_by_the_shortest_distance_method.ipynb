```python
import numpy as np
def mean_class(dir):
    my_file = open(dir, "r")
    data = my_file.read()

    data_into_list = data.replace('\n', ' ').split(" ")
    while ("" in data_into_list):
        data_into_list.remove("")

    a = 5
    lat = []
    long = []

    while a <= len(data_into_list):
        lat.extend([data_into_list[a]])
        long.extend([data_into_list[a + 1]])
        a += 14

    my_file.close()
    List_M_Mean = [np.mean([float(x) for x in lat]), np.mean([float(x) for x in long])]
    return List_M_Mean


baresoil_mean = mean_class('baresoil.txt')
building_mean = mean_class('building.txt')
road_mean = mean_class('road.txt')
vegetation_mean = mean_class('vegetation.txt')

new_baresoil = []
new_building = []
new_road = []
new_vegetation = []
def distance_from_class(dir):
    my_file = open(dir, "r")
    data = my_file.read()

    data_into_list = data.replace('\n', ' ').split(" ")
    while ("" in data_into_list):
        data_into_list.remove("")

    a = 5
    lat = []
    long = []

    while a <= len(data_into_list):
        lat.extend([data_into_list[a]])
        long.extend([data_into_list[a + 1]])
        a += 14

    my_file.close()
    lat = [float(x) for x in lat]
    long = [float(x) for x in long]
    i = 0
    while i < len(lat):
        dist_baresoil = np.sqrt((lat[i] - baresoil_mean[0])**2 + (long[i] - baresoil_mean[1])**2)
        dist_building = np.sqrt((lat[i] - building_mean[0])**2 + (long[i] - building_mean[1])**2)
        dist_road = np.sqrt((lat[i] - road_mean[0])**2 + (long[i] - road_mean[1])**2)
        dist_vegetation = np.sqrt((lat[i] - vegetation_mean[0])**2 + (long[i] - vegetation_mean[1])**2)

        min_dist = [dist_baresoil, dist_building, dist_road, dist_vegetation]
        min_class = min(min_dist)
        if min_class == dist_baresoil:
            non_list = [lat[i], long[i]]
            new_baresoil.extend([non_list])
        elif min_class == dist_building:
            non_list = [lat[i], long[i]]
            new_building.extend([non_list])
        elif min_class == dist_road:
            non_list = [lat[i], long[i]]
            new_road.extend([non_list])
        elif min_class == dist_vegetation:
            non_list = [lat[i], long[i]]
            new_vegetation.extend([non_list])

        i += 1


distance_from_class('baresoil.txt')
distance_from_class('building.txt')
distance_from_class('road.txt')
distance_from_class('vegetation.txt')

print('*************')
print('New baresoil class :\n' + str(new_baresoil))
print('New vegetation class :\n' + str(new_vegetation))
print('New road class :\n' + str(new_road))
print('New building class :\n' + str(new_building))

print('*************')
print('programmer : sajjad hajizade')
print('*************')
```
